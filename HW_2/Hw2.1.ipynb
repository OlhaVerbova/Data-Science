​**Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)" посилання**

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
source = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
df = pd.read_html(source, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]
df
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

region_data = df[df['Регіон'] == 'Київ'].iloc[0, 1:].astype(float)

plt.plot(region_data.index, region_data.values, marker='o', linestyle='-', color='green', label='Київ')
plt.xlabel('Year')
plt.ylabel('Fertility rate')
plt.title('Fertility rate in Kiev (1950-2019)')
plt.legend()
plt.show()
```

```python
import seaborn as sns
sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))
ax = sns.barplot(x="2000", y="Регіон", hue="Регіон", data=df, palette="viridis", dodge=False, legend=False)
plt.title('2000 - Fertility rate in the regions of Ukraine', fontsize=16)
plt.xlabel('2000 - Fertility rate in the regions of Ukraine', fontsize=14)
plt.ylabel('Region', fontsize=14)
plt.show()
```

**Видаліть з таблиці дані по всій країні, останній рядок таблиці**

```python
df =df.drop(df.index[-1])
df
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df=df.fillna(df.mean(numeric_only=True)).round(1)
df
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
df_mean_2019 = df["2019"].mean().round(1)
df_more_than_mean_2019 = df[df["2019"] > df_mean_2019][["Регіон", "2019"]]
print("Average value for 2019:", df_mean_2019)

print("Regions with a 2019 value higher than average:")
print(df_more_than_mean_2019.sort_values(by="2019"))
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
max_born_2014 = df['2014'].max()
max_born_2014_region = df[df['2014']==max_born_2014]
result = max_born_2014_region['Регіон']
print(result.values[0])
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
regions = df["Регіон"]
rate_2019 = df["2019"]


plt.figure(figsize=(12, 6))  
plt.bar(regions, rate_2019, color='skyblue')

max_rate = rate_2019.max()
plt.axhline(max_rate, color='red', linestyle='-', label=f'Max Birth Rate: {max_rate} ‰')

plt.title('Birth rate by region in 2019', fontsize=16)
plt.xlabel('Region', fontsize=14)
plt.ylabel('Fertility rate, ‰', fontsize=14)
plt.xticks(rotation=90) 


plt.tight_layout()  
plt.show()
```
